```python

fig, ax = figax(12,6)
nowdir = '/home/xiaodongli/software/APLike/test/'
for nowfile in ['sdssdr12_AP_lcdm_chisq.mucut0.97.s20to25.table', 
                'sdssdr12_AP_lcdm_chisq.mucut0.97.s20to25.withsys.table']:
    X, Y = XYfromdata(np.loadtxt(nowdir+nowfile)); Y0=[yy/2.0 for yy in Y]
    ymin = min(Y); Y = [yy-ymin for yy in Y]; Y = [exp(-0.5*yy) for yy in Y]
    ax.plot(X,Y, label=nowfile,lw=2)
    data = np.column_stack([Y,Y0,X,[0.7 for yy in Y],[-1 for yy in Y],[0 for yy in Y],[0 for yy in Y]]); 
    nowf=open(nowfile+'.txt','w')
    for row in range(len(data)):
        nowf.write(str(data[row][0])+' '+str(data[row][1])+' '+str(data[row][2])+' '+str(data[row][3])+' '+str(data[row][4])+' '+str(data[row][5])+' '+str(data[row][6])+' \n')
    nowf.close()
    #np.savetxt(nowfile+'.txt',data)   
ax.legend(fontsize=14)   
ax.set_xlim(0.2,0.35)
ax.set_ylim(0,1.3)
ax.set_xlabel(r'$\Omega_m$',fontsize=22)
ax.set_ylabel('Likelihood',fontsize=22)
fig.savefig('AP_om.pdf', format='pdf')
plt.show()
```
